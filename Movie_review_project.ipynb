```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()
```

```python
from nltk import pos_tag
w = "better"
pos_tag([w])
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    if tag.startswith("V"):
        return wordnet.VERB
    if tag.startswith("N"):
        return wordnet.NOUN
    if tag.startswith("R"):
        return wordnet.ADV
    else :
        return wordnet.NOUN
```

```python
documents = [(clean_review(document),category) for document,category in documents]
```

```python
documents[0]
```

```python
training_doc = documents[0:1500]
testing_doc = documents[1500:]
```

```python
#getting all the words in the document
all_words = []
for doc in training_doc:
    all_words+=doc[0]
```

```python
import nltk
```

```python
#it creates a frequency table of all the words
freq = nltk.FreqDist(all_words)
```

```python
#this will most common words out of all the words in the document
common = freq.most_common(3000)
```

```python
#we are storing the most common 3000 words in teh feature list
feature = []
for i in common :
    feature.append(i[0])
```

```python
feature
```

```python
def get_feature_dit(words):
    current_feature = {}
    word_set = set(words)
    for w in feature:
        #this willl give true if w is the wordset else false
        current_feature[w] = w in word_set
    return current_feature
```

```python
get_feature_dit(training_doc[0][0])
```

```python
#creating the dictionary in the form [({f1->true,f2->false ----},category)----]
training_data = [(get_feature_dit(doc),category) for doc,category in training_doc]
testing_data = [(get_feature_dit(doc),category) for doc,category in testing_doc]
```

```python
training_data[0]
```

```python
#using naive bayes from nltk
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
#this will give the most informative features
classifier.show_most_informative_features(15)
```
